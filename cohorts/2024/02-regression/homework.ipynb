```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# Load the dataset from the URL
url = 'https://raw.githubusercontent.com/alexeygrigorev/datasets/master/housing.csv'
data = pd.read_csv(url)
data
```

```python

# Step 2: Filter the data to include only records where 'ocean_proximity' is '<1H OCEAN' or 'INLAND'
filtered_data = data[data['ocean_proximity'].isin(['<1H OCEAN', 'INLAND'])]

# Select the relevant columns
columns = ['latitude', 'longitude', 'housing_median_age', 'total_rooms', 'total_bedrooms',
           'population', 'households', 'median_income', 'median_house_value']
subset_data = filtered_data[columns]
subset_data
```

```python
# Check for missing values
missing_values = subset_data.isnull().sum()
missing_values
```

### Prepare and split the dataset

* Shuffle the dataset (the filtered one you created above), use seed `42`.
* Split your data in train/val/test sets, with 60%/20%/20% distribution.
* Apply the log transformation to the `median_house_value` variable using the `np.log1p()` function.

```python
# Step 1: Shuffle the dataset using seed 42
data_shuffled = subset_data.sample(frac=1, random_state=42).reset_index(drop=True)

# Step 2: Split the dataset into 60% train, 20% validation, and 20% test sets
train_data, temp_data = train_test_split(data_shuffled, test_size=0.4, random_state=42)
val_data, test_data = train_test_split(temp_data, test_size=0.5, random_state=42)

# Step 3: Apply log transformation to the 'median_house_value' column in each set
train_data['median_house_value'] = np.log1p(train_data['median_house_value'])
val_data['median_house_value'] = np.log1p(val_data['median_house_value'])
test_data['median_house_value'] = np.log1p(test_data['median_house_value'])
```
